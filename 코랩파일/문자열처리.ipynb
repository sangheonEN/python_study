```
poem = open("/content/sample_data/poem.txt",'r')
text = poem.read()


print(text)
print(len(text))
word = '.'
print(text.find(word))
print(text.index(word))
print(text.count(word))
print(text[0:text.find(word)])


```

```
dist = text.replace('?', '')
dist1 = dist.split()
dist2 = dist.split('\n')
print(dist1)
print(len(dist1))
print(dist2)
print(len(dist2))
dup = {}
for word in dist1:
  if word not in dup.keys():
    dup[word] = 1 
  else:
    dup[word] += 1

print(dup.keys())
```

```
import matplotlib.pyplot as plt

x = [word for word in dup.keys()]
y = [word for word in dup.values()]
print(x)
print(y)
plt.figure(figsize=(100,15))
plt.bar(x,y, color='b',label='word')
plt.xlabel('Word', fontsize=30, color='w')
plt.ylabel('Count', fontsize=30, color='w')
plt.xticks(x, fontsize=22, color='r')
plt.title('Pome of Word Count',fontsize=50, color='w')
plt.rcParams['axes.grid'] = True
plt.legend(loc=4)
plt.show()
```
