```
rating = open("/content/sample_data/rating1.txt", 'r')
text = rating.read()

a = text.split('\n')
round = len(a)
print(a)
print(round)
labels= a[0].split('\t')
print(labels)
temp = []
docm = []
lab = []
dist = {'id':'', 'document':'','labels':''}

for i in range(1,len(a)):  
   ch = a[i][0:a[i].find('\t')]
   x = a[i][a[i].find('\t'):-1]
   doc = x.strip('\t')
   la = a[i][0-1]
   temp.append(ch)
   docm.append(doc)
   lab.append(int(la))

dist['id'] = temp
dist['document'] = docm  
dist['labels'] = la
   
co = len(temp)
avg = (sum(lab) / co) * 10
count_low = lab.count(0)
count_high = lab.count(1)

   
'''print(temp)
print(docm) 
print(lab)'''

print('영화평을 작성한 인원은 ', co,'입니다.')
print('영화 호감도를 10점 만점으로 환산한 점수는 %.2f 점입니다.' %avg)
print('영화의 호감을 표시한 사람 수는',count_high,'명이며, 비호감을 표시한 사람의 수는 ',count_low,'명입니다.')

#print(dist)
```
